```python
#Adam
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
y=np.array([[0],[1],[1],[0]])
X=np.array([[0,0],[0,1],[1,0],[1,1]])
model=Sequential()
model.add(Dense(16,input_dim=2,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss="mean_squared_error",optimizer="adam",metrics=['binary_accuracy','mean_squared_error'])
history=model.fit(X,y,epochs=1000,verbose=0)
errors=history.history['loss']
y_pred=model.predict(X).round()
num_correct_predictions=(y_pred==y).sum()
accuracy=(num_correct_predictions/y.shape[0])*100
print(accuracy)

```

```python
#Adagrad
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adagrad

y = np.array([[0],[1],[1],[0]])
X = np.array([[0,0],[0,1],[1,0],[1,1]])

model = Sequential()
model.add(Dense(16, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))


model.compile(
    loss="mean_squared_error",
    optimizer=Adagrad(learning_rate=0.5),
    metrics=['binary_accuracy','mean_squared_error']
)

history = model.fit(X, y, epochs=1000, batch_size=2, verbose=0)

errors = history.history['loss']

y_pred = model.predict(X).round()
num_correct_predictions = (y_pred == y).sum()
accuracy = (num_correct_predictions / y.shape[0]) * 100

print(accuracy)
```

```python
#RMSPROP
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop

y = np.array([[0],[1],[1],[0]])
X = np.array([[0,0],[0,1],[1,0],[1,1]])

model = Sequential()
model.add(Dense(16, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

# Changed optimizer to RMSprop
model.compile(
    loss="mean_squared_error",
    optimizer=RMSprop(learning_rate=0.01),
    metrics=['binary_accuracy','mean_squared_error']
)

history = model.fit(X, y, epochs=1000, batch_size=2, verbose=0)

errors = history.history['loss']

y_pred = model.predict(X).round()
num_correct_predictions = (y_pred == y).sum()
accuracy = (num_correct_predictions / y.shape[0]) * 100

print(accuracy)
```
